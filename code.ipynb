```python
import requests
import bs4
import pandas as pd
```

```python
title=[] #先列出想要爬取的欄位
company=[]
region=[]
experience_requirement=[]
education_requirement=[]
job_description=[]
salary=[]
job_update_date=[]
job_url=[]
list1=[region,experience_requirement,education_requirement]
for i in range(1,16): #先人工搜尋有幾頁 (理賠人員職缺共有16頁)
    page=i
    url =f"https://www.104.com.tw/jobs/search/?ro=0&kwop=1&keyword=理賠&expansionType=area,spec,com,job,wf,wktm&order=15&asc=0&page={page}&mode=s&langFlag=0&langStatus=0&recommendJob=1&hotJob=1"
    html_file=requests.get(url)
    ObjSoup=bs4.BeautifulSoup(html_file.text,'html')
    jobs=ObjSoup.find_all('article',class_="b-block--top-bord job-list-item b-clearfix js-job-item")
    for job in jobs:
        title.append(job.get("data-job-name"))
        company.append(job.get("data-cust-name"))  
        for i in range(3):
            list1[i].append(job.find("ul",class_="b-list-inline b-clearfix job-list-intro b-content").find_all("li")[i].text)
        job_description.append(job.find("p",class_="job-list-item__info b-clearfix b-content").text)
        try:
            salary.append(job.find("div",class_="job-list-tag").find("a").text)
        except:
            salary.append("none")
        job_update_date.append(job.find("div",class_="b-block__left").find("span").text)
        job_url.append("https:"+job.find("div",class_="b-block__left").find("a").get("href"))
```

```python
df = pd.DataFrame({'title': title, 'company': company, 'region': region, 
                   'experience_requirement': experience_requirement, 'education_requirement': education_requirement, 
                   'job_description': job_description, 'salary': salary, 'job_update_date': job_update_date,
                   'url': job_url})
```
